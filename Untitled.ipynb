# Data Split to train and test in ML model

```python
x =df.iloc[ : , 1:5].values
y =df['Species']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=30, random_state=8)
```

# Build a Machine Learning model

```python
model = tree.DecisionTreeClassifier(random_state=2)
model = model.fit(x, y)
```

```python
plt.figure(figsize = (15, 10))
tree.plot_tree(model, filled = True)
```

# Training model and Accuracy of model

```python
# Finding the "n_neighbors" value for which the model performs best
n = 0
best_accuracy_so_far = 0
res = []
for k in range(1, 80, 3):
    model = KNeighborsClassifier(n_neighbors = k)
    model.fit(x_train, y_train)
    
    y_pred=model.predict(x_test)
    accuracy=metrics.accuracy_score(y_test, y_pred)
    
    if accuracy > best_accuracy_so_far:
        n = k
        best_accuracy_so_far = accuracy
    res.append(accuracy)
print(res)
print(n, best_accuracy_so_far)
```

```python
# Plotting the Accuracy list
k = [i for i in range(1, 80, 3)]
plt.plot(k, res,'-*', color = "darkolivegreen")
plt.xlabel("k Value")
plt.ylabel("Accuracy")
```

```python
# Creating the best permorming or highest accuracy model
model = KNeighborsClassifier(n_neighbors = n)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy is : {:.3f}".format(accuracy*100),"%" )
```
